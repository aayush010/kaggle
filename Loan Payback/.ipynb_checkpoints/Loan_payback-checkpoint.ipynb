```python
#OUTLIERS
import statsmodels.api as sm
```

```python
predictive_feature = X_train
Y = train_df['loan_paid_back']
predictive_feature = sm.add_constant(predictive_feature)
result = sm.OLS(Y, predictive_feature).fit()
```

```python
print(result.summary())
```

```python
print(type(result))
```

```python
influence = result.get_influence()
print(influence.summary_frame())
std_resid = influence.resid_studentized_external
print(type(std_resid))
outlier_idx = np.where(np.abs(std_resid) > 0)[0]
print(outlier_idx)
#print(type(influence.resid_studentized_external))
#print(std_resid.shape(), std_resid.idxmin(), std_resid.min())
#cooks_d, p_values = influence.cooks_distance
#print(cooks_d, p_values)
```
