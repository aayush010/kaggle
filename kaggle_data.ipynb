```python
import pandas as pd
import zipfile
import os

!pip install kaggle

dataset_name = 'titanic'

from google.colab import files
files.upload("/root/.config/kaggle")

!kaggle competitions download -c {dataset_name} -p ./data
zip_path = f"./data/{dataset_name}.zip"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    print("ðŸ“¦ Files inside ZIP:")
    print(zip_ref.namelist())

dataframes = {}
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    for file in zip_ref.namelist():
        if file.endswith('.csv'):
            with zip_ref.open(file) as f:
                dataframes[file] = pd.read_csv(f)
                print(f"\nâœ… Loaded {file} â€” shape: {dataframes[file].shape}")

for name, df in dataframes.items():
    print(f"\nðŸ“Š Preview of {name}:")
    print(df.head())
```
